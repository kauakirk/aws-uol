```python
from spacy.lang.pt import Portuguese
from spacy.matcher import Matcher


nlp = Portuguese()


documentos = nlp("Você pode ligar para (51) - 9964656570 ou (11) 12344988 ")


matcher = Matcher(nlp.vocab)


padrao = [
    {"ORTH": "("},
    {"SHAPE": "dd"},
    {"ORTH": ")"},
    {"ORTH": "-", "OP": "?"},
    {"IS_DIGIT": True}
]


matcher.add("telefone", [padrao])


matches = matcher(documentos)


for match_id, inicio, fin in matches:
    print(documentos[inicio:fin])
```
