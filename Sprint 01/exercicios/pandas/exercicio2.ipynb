Converter os tipos de dados, de acordo com os tipos listados abaixo.

```python
dataset = dataset.astype({'Year_of_Release': 'int', 'Critic_Count': 'int', 'User_Score': 'float64'})
```

```python
dataset.dtypes
```

Criar uma função para atualizar o atributo User_Score, para que ele fique na mesma escala do atributo Critic_Score (basta multiplicar por 10)
Use apply para atualizar o dataframe

```python
def update(score):
    return score * 10
```

```python
dataset["User_Score"] = dataset['User_Score'].apply(update)
dataset.head()
```

Excluir as seguintes colunas: ["NA_Sales", "JP_Sales", "EU_Sales", "Other_Sales", "Critic_Count", "User_Count"]

```python
dataset.drop(["NA_Sales", "JP_Sales", "EU_Sales", "Other_Sales", "Critic_Count", "User_Count"], axis = 1, inplace = True)
```

Mostrar a contagem de valores para cada coluna categórica.

```python
for column in dataset.columns:
  if dataset[column].dtype == object:
    print('------- ', column, '---------')
    print(dataset[column].value_counts(normalize = True))
    print()
```

```python
dataset["Score"] = (dataset["Critic_Score"] + dataset["User_Score"]) / 2
```
