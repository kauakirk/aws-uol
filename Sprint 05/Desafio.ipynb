```python
def interval_to_mean(value):
    if isinstance(value, str) and '-' in value:
        start, end = map(float, value.split('-'))
        return (start + end) / 2
    return value
amostra['mpg'] = amostra['mpg'].apply(interval_to_mean)
```

## Dados categ√≥ricos
manufacturer: Nome da marca do carro (ex: Toyota, Ford).
model: Modelo espec√≠fico do carro (ex: Corolla, Mustang).
engine: Tipo de motor do carro (ex: 4 cilindros, V6).
transmission: Tipo de transmiss√£o (ex: manual, autom√°tica).
drivetrain: Tipo de tra√ß√£o (ex: tra√ß√£o dianteira, tra√ß√£o integral).
fuel_type: Tipo de combust√≠vel utilizado pelo carro (ex: gasolina, diesel).
exterior_color: Cor externa do carro.
interior_color: Cor interna do carro.
accidents_or_damage: Indica se o carro j√° sofreu algum acidente (1 = sim, 0 = n√£o). Pode ser considerado categ√≥rico ou bin√°rio.
one_owner: Indica se o carro teve apenas um dono (1 = sim, 0 = n√£o). Pode ser considerado categ√≥rico ou bin√°rio.
personal_use_only: Indica se o carro era usado apenas para fins pessoais (1 = sim, 0 = n√£o). Pode ser considerado categ√≥rico ou bin√°rio.
seller_name: Nome do vendedor do carro.


## Dados num√©ricos
year: Ano de fabrica√ß√£o do carro. (Pode ser convertido em uma vari√°vel de idade: 2025 - year)
mileage: Quilometragem do carro (dist√¢ncia percorrida).
mpg: Milhas por gal√£o (medida de efici√™ncia de combust√≠vel).
driver_rating: Avalia√ß√£o do carro feita por outros motoristas.
driver_reviews_num: N√∫mero de avalia√ß√µes feitas por outros motoristas.
seller_rating: Avalia√ß√£o do vendedor (reputa√ß√£o).
price_drop: Redu√ß√£o de pre√ßo em rela√ß√£o ao pre√ßo inicial do carro.
price: Pre√ßo atual do carro (vari√°vel-alvo no caso de regress√£o).

```python
categorical_cols = ['manufacturer', 'model', 'engine', 'transmission', 
                    'drivetrain', 'fuel_type', 'exterior_color', 'interior_color', 
                    'accidents_or_damage', 'one_owner', 'personal_use_only', 'seller_name']

numerical_cols = ['year', 'mileage', 'mpg', 'driver_rating', 'driver_reviews_num', 
                   'price']
```

### Usando o label encondedr para transformar as colunas categoricas

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
amostra_encoded = amostra.copy()
```

```python
for col in categorical_cols:
    amostra_encoded[col] = label_encoder.fit_transform(amostra_encoded[col].astype(str))
```

```python
amostra_encoded
```

fazendo o primeiro modelo usando a Regressao linear para entender a base se √© simples 


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
correlation_matrix = amostra_encoded.corr()


plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de Correla√ß√£o')
plt.show()
```

As variaveis mais importantes usando o label sao mileage e year

```python
X = amostra_encoded.drop(columns=['price'])
y = amostra_encoded['price']  # Vari√°vel alvo
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
```

```python
y_pred = linear_model.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Erro M√©dio Absoluto (MAE): {mae:.2f}")
print(f"Erro Quadr√°tico M√©dio (MSE): {mse:.2f}")
print(f"Coeficiente de Determina√ß√£o (R¬≤): {r2:.2f}")
```

### Usando o TargetEncoder para Transformar os Dados

```python
from category_encoders import TargetEncoder
target_encoder = TargetEncoder(cols=categorical_cols)
amostra_encoded = amostra.copy()
amostra_encoded[categorical_cols] = target_encoder.fit_transform(amostra[categorical_cols], amostra['price'])
```

```python
correlation_matrix = amostra_encoded.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de Correla√ß√£o')
plt.show()
```

### Ao aplicar o TargetEncoder para transformar as vari√°veis categ√≥ricas, as vari√°veis model e engine se destacaram como as mais importantes para o modelo

Usando o random forest para construir modelo

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
X = amostra_encoded.drop(columns=['price'])
y = amostra_encoded['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
rf_model = RandomForestRegressor(random_state=42)
```

```python
rf_model.fit(X_train, y_train)
```

```python
y_pred = rf_model.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Erro M√©dio Absoluto (MAE): {mae:.2f}")
print(f"Erro Quadr√°tico M√©dio (MSE): {mse:.2f}")
print(f"Coeficiente de Determina√ß√£o (R¬≤): {r2:.2f}")
```

### O resultado do modelo foi muito boa porem vou testar mais um antes de realziar o tuning

## Usando gradient boosting

```python
from sklearn.ensemble import GradientBoostingRegressor

gb_model = GradientBoostingRegressor(random_state=42)
gb_model.fit(X_train, y_train)

y_pred_gb = gb_model.predict(X_test)

mae_gb = mean_absolute_error(y_test, y_pred_gb)
mse_gb = mean_squared_error(y_test, y_pred_gb)
r2_gb = r2_score(y_test, y_pred_gb)

print(f"Erro M√©dio Absoluto (MAE): {mae_gb:.2f}")
print(f"Erro Quadr√°tico M√©dio (MSE): {mse_gb:.2f}")
print(f"Coeficiente de Determina√ß√£o (R¬≤): {r2_gb:.2f}")
```

### o desempenho nao foi melhor entao vamos continuar com o random forest

### fazendo tuning no random forest usando o randomized search

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_dist = {
    'n_estimators': [100, 200, 300, 500],  # Testar diferentes n√∫meros de √°rvores
    'max_depth': [10, 20, 30, None],        # Testar diferentes profundidades
    'min_samples_split': [2, 5, 10],        # Testar diferentes n√∫meros de amostras para dividir um n√≥
    'min_samples_leaf': [1, 2, 4],          # Testar diferentes n√∫meros de amostras em uma folha
    'bootstrap': [True, False]              # Testar com e sem bootstrap
}
rf = RandomForestRegressor(random_state=42)
```

```python
random_search = RandomizedSearchCV(estimator=rf, param_distributions=param_dist, n_iter=100, cv=3, n_jobs=-1, random_state=42, verbose=2)
```

```python
random_search.fit(X_train, y_train)
```

```python
print(f"Melhores par√¢metros: {random_search.best_params_}")
best_rf_model = random_search.best_estimator_
y_pred_rf = best_rf_model.predict(X_test)
```

```python
mae_rf = mean_absolute_error(y_test, y_pred_rf)
mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)
print(f"Erro M√©dio Absoluto (MAE): {mae_rf:.2f}")
print(f"Erro Quadr√°tico M√©dio (MSE): {mse_rf:.2f}")
print(f"Coeficiente de Determina√ß√£o (R¬≤): {r2_rf:.2f}")
```
